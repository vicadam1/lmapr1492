Il nous est demandé de déterminer les 3 premiers pics du diffractogramme de notre cristal et de préciser les indices hkl correspondants. On considère dans notre cas un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Angström. Sur base du modèle de diffraction, nous pouvons trouver la position de chacun des atomes du cristal.


```python
from pymatgen.ext.matproj import MPRester

#Création de la structure du cristal à partir du MP
with MPRester ("m5wyacFUOgt3WAfAgu9M") as m:
    structure = m.get_structure_by_material_id("mp-867168")
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator

xrd = XRDCalculator(1.54060)
pat = xrd.get_pattern(structure)
```

Ci-dessous, nous pouvons observer le diffractogramme du cristal. Celui-ci représente l'angle 2$\theta$ en degré en fonction de l'intensité mise à l'échelle.

```python
#Affichage du diffractogramme 
xrd.show_plot(structure,two_theta_range=(0,45))
```

# Premier pic


```python
theta1 = pat.x[0]
I1 = pat.y[0]
Indices_hkl1 = pat.hkls[0]

#print
print('Premier pic')
print('-----------')
print('Angle de diffraction : '+str(theta1)+' degrés')
print('Intensité : '+str(I1))
print('Indices hkl : '+str(Indices_hkl1))
```

On peut voir que le premier pic se trouve à un angle 2𝜃 = 19.356184464424146 degrés et a une intensité de 4.907416858451089. Ce que nous pouvons vérifier sur notre graphe et qui semble correcte. Les indices hkl correspondants sont (1, 0, 0) de multiplicité 6 et (1, 1, 1) de multiplicité 2. 

# Deuxième pic


```python
theta2 = pat.x[1]
I2 = pat.y[1]
Indices_hkl2 = pat.hkls[1]

#print
print('Deuxième pic')
print('-----------')
print('Angle de diffraction : ' + str(theta2) + ' degrés')
print('Intensité : ' + str(I2))
print('Indices hkl : ' + str(Indices_hkl2))
```

On peut voir que ce deuxième pic se trouve à un angle 2𝜃 = 22.3865907641368 degrés et a une intensité de 9.623856163213343. Cela semble également correcte quand nous comparons ces résultats avec le graph. Les indices hkl correspondants sont (1, 1, 0) et la multiplicité est de 6.

# Troisième pic

```python
theta3 = pat.x[2]
I3 = pat.y[2]
Indices_hkl3 = pat.hkls[2]

#print
print('Troisième pic')
print('-----------')
print('Angle de diffraction : ' +str(theta3)+' degrés')
print('Intensité : ' +str(I3))
print('Indices hkl : ' +str(Indices_hkl3))
```

On peut voir que le pour ce troisième pic, l'angle est de 2𝜃 = 31.86759916179437 degrés et l'intensité est de 100. L'intensité sur le graph correspond a cette affirmation. Les indices hkl correspondants sont (1, 0, -1) de multiplicité 6 et (2, 1, 1) également de multiplicité 6.
